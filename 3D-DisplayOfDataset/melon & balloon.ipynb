```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
```

```python
melon_width = np.random.normal(40, 20 ,N)
melon_height = np.random.normal(20, 10 ,N)
melon_weight = np.random.normal(7000, 300 ,N)

balloon_width = np.random.normal(20, 2, N)
balloon_height = np.random.normal(20, 2, N)
balloon_weight = np.random.normal(350, 30, N)

width = np.concatenate((melon_width, balloon_width))
height = np.concatenate((melon_height, balloon_height))
weight = np.concatenate((melon_weight, balloon_weight))

X = np.array([width, height, weight]).T
melon_label = np.zeros(N, dtype= 'int')
balloon_label = np.ones(N, dtype= 'int')

Y = np.concatenate((melon_label, balloon_label))


```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

ax.scatter(X[:N, 0], X[:N ,1], X[:N ,2], c = 'orange', marker='^')
ax.scatter(X[N: , 0], X[N: ,1], X[N: ,2], c = 'blue', marker='o' )

ax.set_xlabel('width(cm)')
ax.set_ylabel('height(cm)')
ax.set_zlabel('weight(gr)')
```
